```python
# Importar libreras necesarias
import numpy as np
import matplotlib.pyplot as plt

# Example 6.1.1: Monte Carlo Integration
def monte_carlo_integration(f, a, b, N):
    x = np.random.uniform(a, b, N)
    integral = (b - a) * np.mean(f(x))
    return integral

# Funcin a integrar
f = lambda x: x**2
a, b = 0, 1  # Lmites de integracin
N = 10000  # Nmero de puntos
result = monte_carlo_integration(f, a, b, N)
print(f"Monte Carlo Integration result: {result}")

# Problema 6-0
x = np.linspace(0, 5, 500)
y1 = x**4 * np.exp(-2 * x)
y2 = (x**2 * np.exp(-x) * np.sin(x**2))**2

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r"$x^4 e^{-2x}$")
plt.plot(x, y2, label=r"$(x^2 e^{-x} \sin(x^2))^2$")
plt.title("Graphs of Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Problema 6-2
# Simular lectura del archivo Ba137.txt (crearemos datos de ejemplo)
time = np.linspace(0, 10, 100)
counts = 100 * np.exp(-0.3 * time) + np.random.normal(0, 5, len(time))

# Grfico de los datos
plt.figure(figsize=(10, 6))
plt.plot(time, counts, 'bo', label="Experimental Data")
plt.title("Counts vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Counts")
plt.legend()
plt.grid(True)
plt.show()

# Verificacin de comportamiento logartmico
ln_counts = np.log(counts)
plt.figure(figsize=(10, 6))
plt.plot(time, ln_counts, 'ro', label="ln(Counts)")
plt.title("Logarithmic Transformation of Data")
plt.xlabel("Time (s)")
plt.ylabel("ln(Counts)")
plt.legend()
plt.grid(True)
plt.show()
```
